```python
def load_page(url):
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
    except Exception as e:
        print(e)
        return
    return soup
soup = load_page('https://nips.cc/Conferences/2017/Schedule')
```

```python
assert type(soup) == BeautifulSoup
```

```python
cards = soup.find_all(class_='maincardFooter')
```

```python
def split_text(text, character=[".", chr(183)]):
    try:
        new_names = text.split(character[0])
    except ValueError as e:
        print(text)
    if len(new_names) == 1 and len(character) > 1:
        new_names = split_text(text, character[1:])
    return new_names
split_text('Nando de Freitas · Scott Reed · Oriol Vinyals')
```

```python
names = []
for card in cards:
    new_names = split_text(card.text)
    if new_names:
        names.extend([name.strip() for name in new_names])
len(names)
```

```python
def drop_duplicates(mylist):
    return list(set(mylist))
names = drop_duplicates(names)
len(names)
```

```python
def find_url_str(group):
    try:
        return group("a")[0].get("href")
    except IndexError:
        return
```

```python
def find_links(r):
    groups = r.find_all(class_='g')
    pages = []
    for group in groups:
        url_str = find_url_str(group)
        if url_str:
            pages.append(parse_url_str(url_str))
    return pages
links = find_links(r)
```

```python
def retrieve_links_for_name(name):
    google_url = 'https://www.google.com/search?q='
    name_search = name.replace(" ", "+")
    if len(name_search) < 1:
        return
    page_soup = load_page(google_url + name_search + "+statistics")
    links = find_links(page_soup)
    return links
links = retrieve_links_for_name(names[3])
len(links)
```

```python
links
```

```python
def people_search(names):
    data = []
    for name in names:
        row = [name]
        if len(name) > 1:
            links = retrieve_links_for_name(name)
            if links and len(links) > 0:
                row.extend(links)
                sleep(1)
            data.append(row)
    return data
data = people_search(names[:40])
len(data)
```

```python
import pandas as pd
```

```python
people_list = pd.DataFrame(data)
```

```python
people_list.to_csv(os.path.join(DATA_DIR, 'people_list.csv'), index=False)
```
